```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_excel('/content/DoctorVisits.xlsx')
```

```python
print("Dimensions: ", df.shape)
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df = df.fillna(df.mean())
```

```python
plt.scatter(df['age'], df['illness'])
plt.xlabel('Age')
plt.ylabel('Illness')
plt.title('Age vs. Illness')
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Visits')
plt.title('Gender vs. Visits')
plt.show()
```

```python
age_avg_income = df.groupby('age')['health'].mean()
plt.plot(age_avg_income.index, age_avg_income.values)
plt.xlabel('Age')
plt.ylabel('Health')
plt.title('Age vs. Health')
plt.show()
```

```python
dataplot = sb.heatmap(df.corr(), cmap="YlGnBu", annot=True)
plt.show()
```
